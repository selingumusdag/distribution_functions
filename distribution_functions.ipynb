```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
historical_prices = pd.read_csv('data/GOOG.csv', index_col=0)
#historical_prices
close_price=historical_prices["Close"]
#close_price
x=1
stock_list=[]

while x<len(close_price)-1:
    previous_day=close_price[x-1]
    today=close_price[x]
    stock=(today-previous_day)/previous_day
    stock_list.append(stock)
    x+=1
#print(stock_list)
normalization_data=np.random.normal(np.mean(stock_list),np.std(stock_list),len(stock_list))

stock_list.sort()
normalization_data.sort()

plt.figure(figsize=(8,4),dpi=100)
title_font={'family':'arial','color':'red','weight':'bold','size':15}
axis_font={'family':'arial','color':'black','weight':'bold','size':13}
plt.title('Stock Prices of Google',fontdict=title_font)
plt.ylabel('Stock Price of Normalizaton', fontdict = axis_font) 
plt.xlabel('Stock Price', fontdict = axis_font)
plt.scatter(normalization_data, stock_list)
plt.plot([np.min(normalization_data), np.max(normalization_data)],
         [np.min(stock_list), np.max(stock_list)], c='red')
plt.show
```
